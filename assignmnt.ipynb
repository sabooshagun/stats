```python
import pandas as pd
```

```python
df=pd.read_csv('/Users/shagunsaboo/Downloads/startup.csv')
df.head()
```

```python
delhi_val = (df.loc[(df["CityLocation"]=="New Delhi") | (df["CityLocation"]=="Banglore") | (df["CityLocation"]=="banglore") | (df["CityLocation"]=="delhi")] )
```

```python
delhi_val
```

```python
amount_city=delhi_val['AmountInUSD']
amount_city.head()
                      
```

```python
amount_city.dropna(inplace=True)
amount_city.head()

a1=amount_city.str.replace(',','')
a1.head()
```

```python
a2=pd.to_numeric(a1)
a2
```

```python
pop_avg=a2.mean()
pop_min=a2.min()
pop_max=a2.max()
print(pop_avg)
print(pop_min)
print(pop_max)
```

```python
population=a2
sample_size=75
sample=population.sample(sample_size,random_state = 1)
sample.shape
```

```python
sample_avg=sample.mean()
sample_min=sample.min()
sample_max=sample.max()
print(pop_avg," ",sample_avg)
print(pop_min," ",sample_min)
print(pop_max," ",sample_max)
```

```python
sampling_error=pop_avg-sample_avg
print(sampling_error)
```
